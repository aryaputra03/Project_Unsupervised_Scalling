```python
x_mm = df_mm.drop('AveragePrice',axis=1)
y_mm = df_mm['AveragePrice']
```

```python
x_ss = df_ss.drop('AveragePrice',axis=1)
y_ss = df_ss['AveragePrice']
```

```python
x_rob_col = x_rob.columns
```

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
x_rob = pd.DataFrame(rs.fit_transform(x_rob),columns=x_rob_col) 
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

mm = MinMaxScaler()

x_mm = pd.DataFrame(mm.fit_transform(x_mm),columns=xcol) 
```

```python
std = StandardScaler()
x_ss = pd.DataFrame(std.fit_transform(x_ss),columns=xcol)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
```

# No Scale

```python
# cv_no_scale = cross_val_score(linreg,X=x_no_scale,y=y_no_scale,cv=25,scoring=['neg_mean_squared_error'])
models = {
    "Regresi": LinearRegression(),
    "KNN": KNeighborsRegressor(),
    "SVM": SVR(),
    "RF": RandomForestRegressor(),
    "Cat": CatBoostRegressor(verbose=0)
}
hasil={}
for i,j in models.items():
    cv_no_scale = cross_val_score(j,X=x_no_scale,y=y_no_scale,cv=25,scoring='neg_mean_squared_error')
    # print(f"{i} = {cv_no_scale}")
    hasil[i] = cv_no_scale
hasil_no_scale = pd.DataFrame(hasil)

```

```python
abs(hasil_no_scale).to_csv('data_1_hasil_no_scale.csv',index=False)
```

# MM

```python

hasil1={}
for i,j in models.items():
    cv_mm = cross_val_score(j,X=x_mm,y=y_mm,cv=25,scoring='neg_mean_squared_error')
    # print(f"{i} = {cv_mm}")
    hasil1[i] = cv_mm
hasil_mm = pd.DataFrame(hasil1)

```

```python
abs(hasil_mm).to_csv('data_1_hasil_Minmax.csv',index=False)
```

# SS

```python
hasil2={}
for i,j in models.items():
    cv_ss = cross_val_score(j,X=x_ss,y=y_ss,cv=25,scoring='neg_mean_squared_error')
    # print(f"{i} = {cv_ss}")
    hasil2[i] = cv_ss
hasil_ss = pd.DataFrame(hasil2)
```

```python
abs(hasil_ss).to_csv('data_1_hasil_Std.csv',index=False)
```

# Robust

```python
hasil3 = {}

for i,j in models.items():
    cv_rob = cross_val_score(j,X=x_rob,y=y_rob,cv=25,scoring='neg_mean_squared_error')
    # print(f"{i} = {cv_rob}")
    hasil3[i] = cv_rob
hasil_rob = pd.DataFrame(hasil3)
```

```python
abs(hasil_rob).to_csv('data_1_hasil_rob.csv',index=False)
```
